Deaths among states

```python
states = np.array(data['State/UTs'])
deaths = np.array(data['Deaths'])

ndf = pd.DataFrame({'states':states,
                    'deaths':deaths})

resetidx = ndf.set_index('states')
sorteddf = resetidx.deaths.sort_values(ascending=False).head()
print("States with highest number of deaths: ",sorteddf)

sorteddf.plot(kind='bar', x='states', y='deaths',
             title='Largest number of deaths among all states',
             figsize=(12,6),
             legend=True,
             xlabel="States",
             ylabel="Number of Deaths",
             fontsize=10,
             color="blue",
             rot=1)
plt.show
```
